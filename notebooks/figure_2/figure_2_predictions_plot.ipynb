```python
import h5py
import matplotlib.animation as ani
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import cm
import os

DATA_DIR = "/media/data/merkourios/data_for_figs"
```

```python
flag3d_ground_truth = np.load(os.path.join(DATA_DIR, "flag_5_ground_truth.npy")).T
flag3d_predictions = np.load(os.path.join(DATA_DIR, "flag_5_prediction.npy")).T
flag3d_spindles = np.load(os.path.join(DATA_DIR, "flag_5_spindles.npy"))
flag3d_shoulder = np.load(os.path.join(DATA_DIR, "flag_shoulder_5.npy")).T
flag3d_elbow = np.load(os.path.join(DATA_DIR, "flag_elbow_5.npy")).T
flag3d_wrist = np.load(os.path.join(DATA_DIR, "flag_wrist_5.npy")).T
flag3d_endeffector_coords = flag3d_wrist
```

```python
flag3d_ground_truth = np.load(os.path.join(DATA_DIR, "flag_5_ground_truth.npy")).T
flag3d_predictions = np.load(os.path.join(DATA_DIR, "flag_5_prediction.npy")).T

ground_truth = flag3d_ground_truth[(6, 0, 1, 2), :]
predictions = flag3d_predictions[(6, 0, 1, 2), :]
time = np.linspace(0, ground_truth.shape[1] / 240, ground_truth.shape[1])


# Labels for subplots
subplot_titles = ['Angle (Degrees)', 'Distance (cm)', 'Distance (cm)', 'Distance (cm)']

# Create the figure and subplots
fig, axes = plt.subplots(2, 2, figsize=(8, 6), dpi=300)
fig.subplots_adjust(hspace=0.4, wspace=0.3)  # Adjust spacing between subplots

# Plot each ground truth and prediction
for i, ax in enumerate(axes.flat):
    # Plot ground truth
    ax.plot(time, ground_truth[i], label='Ground Truth', color='blue', linewidth=1.5)
    # Plot prediction
    ax.plot(time, predictions[i], label='Prediction', color='orange', linewidth=1.5)
    
    # Add labels, title, and grid
    ax.tick_params(axis='both', labelsize=7)
    ax.grid(visible=True, linestyle='--', linewidth=0.5, alpha=0.7)
    
    # Remove top and right spines
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)

# Show the plot
plt.show()
```
