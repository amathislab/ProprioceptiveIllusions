# Figure 2, Panel C (Middle)

```python
import h5py
import matplotlib.animation as ani
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import cm
import os
from matplotlib.colors import to_rgba
# from directory_paths import SAVE_DIR, PARENT_DIR

DATA_DIR = "/media/data/merkourios/data_for_figs"
```

```python
flag3d_spindles = np.load(os.path.join(DATA_DIR, "flag_5_spindles.npy"))
pcr_spindles = np.load(os.path.join(DATA_DIR, "rand_89_spindles.npy"))
```

```python
triceps_color = "00C7BE" # Hex value
biceps_color = "FF3B30" # Hex value

triceps_color = [int(triceps_color[i:i+2], 16) / 255 for i in (0, 2, 4)]
biceps_color = [int(biceps_color[i:i+2], 16) / 255 for i in (0, 2, 4)]
```

## FLAG3D

```python
biceps_spindles = flag3d_spindles[:, 16, :]
triceps_spindles = flag3d_spindles[:, 24, :]

for neuron_data, color in zip([biceps_spindles, triceps_spindles], [biceps_color, triceps_color]):
    time = np.linspace(0, neuron_data.shape[1] / 240, neuron_data.shape[1])

    # Base color and y-offset
    base_color = color
    y_offset = 10  # Distance between each neuron

    # Create the figure and axis
    fig, ax = plt.subplots(figsize=(6, 4), dpi=300)  # High resolution for publication

    # Plot each neuron's data
    for i, data in enumerate(neuron_data):
        # Adjust color saturation (alpha channel)
        color = to_rgba(base_color, alpha=0.5 + 0.5 * (i / len(neuron_data)))
        # Apply y-offset
        ax.plot(time, data + i * y_offset, color=color, linewidth=2)

    # Adjust tick parameters for better visibility
    ax.tick_params(axis='both', which='major', labelsize=15)
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)

    # Add gridlines for clarity
    ax.grid(visible=True, which='both', linestyle='--', linewidth=0.5, alpha=0.7)

    # Optimize layout
    plt.tight_layout()

    # Show the plot
    plt.show()
```
